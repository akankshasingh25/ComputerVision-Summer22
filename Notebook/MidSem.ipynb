```python
import numpy as np

# The intersection of two lines in Homogeneous System
# is the matrix product of the two lines

L1 = np.asmatrix([1, 1, -5])
L1 = L1.T
print(L1)

L2 = np.asmatrix([4, -5, 7])
L2 = L2.T
print(L2)

Intersection_HC = np.asarray([((L1[1]*L2[2])-(L1[2]*L2[1])),
                            ((L1[2]*L2[0])-(L1[0]*L2[2])),
                            ((L1[0]*L2[1])-(L1[1]*L2[0]))
                           ]).reshape(3)
print(Intersection_HC)

w = -9.0
Intersection_EC = Intersection_HC/w
print(Intersection_EC)
Intersection_EC = np.delete(Intersection_EC, 2, 0)
print(Intersection_EC)
```

```python
R = np.asmatrix([[0.9, 0.4, 0.1732],
     [-0.4183, -0.9043, -0.0854],
     [-0.1225, -0.1493, -0.9812,]])
print(R)

COR = np.asmatrix([-1, -2, -3])
COR = COR.T
print(COR)

I = np.identity(3)
print(I)

COR_34 = np.append(I, -COR, 1)
print(COR_34)

M  = np.append(R, -COR, 1)
print(M)

#floating point error
M1  = np.matmul(R, -COR_34)
print(M1)
```

```python
import skimage.io
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

image = cv2.imread('shapes.png')
ret, binary = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY)

#Display graphics
titles = ['GrayScale', 'Binary']
images = [image, binary]
plt.figure(figsize=(12,12))

for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.tight_layout()
plt.show()
```

```python
con_components = cv2.connectedComponents(binary, 8)[1]
```

```python
kernel = np.ones((5,5),np.uint8)
dilation1 = cv2.dilate(binary, kernel, iterations = 1)

plt.imshow(dilation1, 'gray')
plt.title('Dilation')
plt.xticks([]), plt.yticks([])
    
plt.tight_layout()
plt.show()
```
